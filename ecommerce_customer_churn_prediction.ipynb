## Business Problem <a name="#business-problem"></a>

<div id='#business-problem' style="font-size: 21px; font-weight: bold;">1.1 Business Background</div>

<div style="text-align: justify;">
PT. E-Shop is a technology company based in the UK specializing in e-commerce products. They sell a variety of items ranging from electronics, clothing, household appliances, to daily necessities. Over the past few years, PT. E-Shop has experienced significant growth in customer base and transaction volume. However, the company faces challenges in customer retention, especially amidst increasing competition.
<br><br>
The company achieves profitability from customers who make repeat purchases on the e-commerce platform. Therefore, retaining loyal customers is crucial for the company's development and stability. To achieve this, PT. E-Shop focuses on acquiring new customers for business growth and preventing churn, where customers discontinue using the services provided by the company.
</div>


<div id='#business-problem' style="font-size: 21px; font-weight: bold;">1.2 Problem Statement</div>

<div style="text-align: justify;">

Based on statistics across various business industries, research has found that customer acquisition costs 4-5 times more than customer retention. Furthermore, a mere 5% increase in customer retention can lead to at least a 25% increase in profitability. This is because returning customers are likely to spend 67% more on your company's products and services. As a result, your company can reduce operational costs associated with acquiring new customers. [(Source)](https://www.ngdata.com/what-is-customer-churn/)

</div>

<div style="text-align: justify;">

In conclusion, acquiring new customers incurs higher costs compared to retaining existing customers to prevent churn. Therefore, companies must take appropriate actions to prevent and minimize customer churn by providing necessary treatments to ensure customers remain loyal to the company's e-commerce platform. However, these actions must also be targeted and strategic.[(Source)](https://blog.hubspot.com/service/what-is-customer-churn#:~:text=Well%2C%20it%27s%20important,made%20that%20decision.)

</div>


<div id='#business-problem' style="font-size: 21px; font-weight: bold;">1.3 Goals</div>

<div style="text-align: justify;">

Based on the challenges faced, it is crucial for the company to identify customers who are likely to churn in order to take preventive actions to maintain its stability. In this regard, the analysis is conducted with the following objectives:

- Building a model that can predict and analyze customers who have the potential to churn.
  
- Minimizing the retention costs required for customers who intend to churn.
  
- Understanding the characteristics of customers likely to churn and identifying significant factors influencing their churn.
  
- Providing appropriate treatment to these customers to prevent churn.

</div>


<div id='#business-problem' style="font-size: 11px; font-weight: bold;">

# **1.4 Analytical Approach**

</div>

<div style="text-align: justify;">

In this analysis process, the analysis is carried out using a machine learning approach to build a classification model that can predict the probability of customers who may churn. Selecting the right metric is crucial to achieving the established goals.

</div>

Target :

- 0 : Customer not churn
- 1 : Customer Churn

Confussion Metric :

- True Positive (TP) : Customer actually churn and predicted as churn
- True Negative (TN) : Customer actually not churn and predicted as not churn
- False Negative (FN) : Customer actually churn but predicted as not churn
- False Positive (FP) : Customer actually not churn but predicted as churn

Prediction Error :

| **Error Type**     |**Meaning** | **Outcome** | 
|-----------------|------------|----------------|
| **False Positive / Type 1 Error**  | Happens when model incorrectly predicts that the customer will **churn**, when actually not | The company wasting retaining-cost because giving a promo to customers who are already loyal|
| **False Negative / Type 2 Error**  | Happens when the model incorrectly predicts that the customer will **not churn** when actually they will | The company will lose recurring revenue and spends more funds and time to acquire new customers | 


Metric Selection :

<div style="text-align: justify;">

In order to accurately predict customers who will truly churn, the model should prioritize by minimizing False Negatives to prevent the loss of customers who actually churn. By considering that acquiring new customers are more costly than retaining loyal ones, the primary focus should be on recall. Recall measures the model's ability to correctly identify all actual churned customers, so the company could take a decision by putting the cost of retaining customer to whom does it actually needed which is crucial in preventing potential business losses. Therefore, the most suitable metric score to emphasize in this context is the F2 Score. [Source ](https://iopscience.iop.org/article/10.1088/1742-6596/1918/4/042002/pdf)

</div>

<div id='#business-problem' style="font-size: 20px; font-weight: bold;">2.5.2 Categorical Features Correlation</div>

```python
# Melakukan uji chi square
chi2_result = []

for i in categorical_features:
  df_chi = pd.crosstab(df_raw[i], df_raw['Churn'])
  stat, pvalue, dof, expected = chi2_contingency(df_chi)
  chi2_result.append(round(pvalue, 5))

df_chi_result = pd.DataFrame({'categorical_features': categorical_features,'chi2_score': chi2_result})
df_chi_result['notes'] = np.where(df_chi_result['chi2_score'] > 0.05, 'independent', 'dependent')
df_chi_result
```

<div style="text-align: justify;">

Based on the results of the chi-square test above, it is known that all categorical features have a chi-square score < 0.05 (successfully rejecting Ho). This means that each categorical feature is significantly associated with the target variable Churn (the proportions differ significantly).

<div id='#outliers' style="font-size: 20px; font-weight: bold;">4.1.3 Outliers</div>

```python
# checking outliers using boxplot for the continuous variable

continuous_variables = ['Tenure', 'WarehouseToHome', 'DaySinceLastOrder', 'CashbackAmount']
plt.figure(figsize=(15,20))
for i in range(0, len(continuous_variables)):
  plt.subplot(4, 4, i+1)
  sns.boxplot(y=df_raw[continuous_variables[i]], color='blue', orient='v')
  plt.tight_layout()
```

```python
# Initialize an empty list to store the results
results = []

# Loop through each continuous variable
for i in range(len(continuous_variables)):
    Q1 = df_raw[continuous_variables[i]].quantile(0.25)
    Q3 = df_raw[continuous_variables[i]].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    condition = (df_raw[continuous_variables[i]] < lower_bound) | (df_raw[continuous_variables[i]] > upper_bound)
    outlier = df_raw.loc[condition, continuous_variables[i]].count()
    PersentaseOutlier = outlier / df_raw[continuous_variables[i]].count() * 100
    
    # Append the results to the list
    results.append({
        'Variable': continuous_variables[i],
        'Lower Bound': round(lower_bound, 2),
        'Upper Bound': round(upper_bound, 2),
        'Outliers': outlier,
        'Percentage Outliers': round(PersentaseOutlier, 2) 
    })

# Create a DataFrame from the results list
df_outliers = pd.DataFrame(results)

# Display the DataFrame
df_outliers
```

From the calculations above, outliers in the `Tenure` and `WarehouseToHome` columns will be dropped due to their small percentage in the Data Cleaning process. Meanwhile, outliers in the `DaySinceLastOrder` and `CashbackAmount` columns will be ignored as they are considered reasonable values despite having high variance.

<div id='#model-benchmark-evaluation' style="font-size: 20px; font-weight: bold;">6.1 Finding Best Resampling Method</div>


<div style="text-align: justify;">

In this machine learning process, the analyst will compare several different resampling methods against the top 3 models from the benchmark results. This aims to find the best approach to address the imbalanced data issue, where the positive class has a proportion of 16,7% and the negative class has 83.3%. It's crucial to balance this so that the model learns not to bias towards capturing more of the negative class cases. Below are the resampling methods to be used.

</div>


```python
smote = SMOTE(random_state = RANDOM_STATE)
adasyn = ADASYN(random_state = RANDOM_STATE)
svm_smote = SVMSMOTE(random_state = RANDOM_STATE)
borderline_smote = BorderlineSMOTE(random_state = RANDOM_STATE)
ros = RandomOverSampler(random_state = RANDOM_STATE, sampling_strategy= 'minority')
smote_enn = SMOTEENN(smote=SMOTE(random_state = RANDOM_STATE), enn=EditedNearestNeighbours())
```

```python
resampling_techniques = [
    ('ROS', ros), 
    ('SMOTE', smote), 
    ('ADASYN', adasyn), 
    ('SMOTEENN', smote_enn),
    ('SVMSMOTE', svm_smote), 
    ('BorderlineSMOTE', borderline_smote)
]
```

<div id='#xgboost-resampling' style="font-size: 20px; font-weight: bold;">6.1.1 XGBoost Classifier Best Resampling</div>


```python
# Initialize lists to store results
resampling_names = []
xgb_resample_train_score = []
xgb_resample_test_score = []
xgb_resample_score_diff = []

# Loop through each resampling method and each model
for name, resample_method in resampling_techniques:
    model_pipeline = ImbPipeline([
        ('transformer', transformer),
        ('resampling', resample_method),
        ('model', xgb)
    ])
    
    # Fitting the model
    model_pipeline.fit(X_train, y_train)
    
    # Predicting on the training set
    y_pred_train = model_pipeline.predict(X_train)
    
    # Predicting on the test set
    y_pred_test = model_pipeline.predict(X_test)
    
    # Evaluate the scores
    f2_score_train = fbeta_score(y_train, y_pred_train, beta=2)
    f2_score_test = fbeta_score(y_test, y_pred_test, beta=2)
    
    # Calculate the score difference
    score_diff = f2_score_train - f2_score_test
    
    # Append results to lists
    resampling_names.append(name)
    xgb_resample_train_score.append(f2_score_train)
    xgb_resample_test_score.append(f2_score_test)
    xgb_resample_score_diff.append(score_diff)
```

```python
hyperparam_space = {
    'model__boosting_type': ['gbdt', 'dart', 'goss'],
    'model__max_depth': list(np.arange(1, 31)),
    'model__num_leaves': list(np.arange(2, 900, 5)),
    'model__min_data_in_leaf': list(np.arange(10, 101, 2)),
    'model__num_iterations': list(np.arange(10, 101, 2)),
    'model__learning_rate': uniform(0.01, 0.99),
    'model__subsample': uniform(0.5, 0.5),  # Range from 0.5 to 1.0
    'model__colsample_bytree': uniform(0.5, 0.5),  # Range from 0.5 to 1.0
    'model__reg_alpha': uniform(0, 1),  # L1 regularization
    'model__reg_lambda': uniform(0, 1)  # L2 regularization
}

hyperparam_space_ada = {
    'model__n_estimators': [int(x) for x in np.linspace(start=50, stop=200, num=4)],
    'model__learning_rate': [0.01, 0.1, 0.5, 1],
    'model__algorithm': ['SAMME', 'SAMME.R']
}
```

<div id='#tuning-xgboost' style="font-size: 20px; font-weight: bold;">6.2.1 Tuning XGBoost Classifier</div>


```python
# Define your resampling techniques
best_resamples = [
    ('ROS', ros),
    ('SMOTEENN', smote_enn),
    ('ADASYN', adasyn),
    ('SMOTE', smote),
]

# Initialize lists to store results
resampling_names = []
xgb_tuning_train_score = []
xgb_tuning_test_score_before = []
xgb_tuning_test_score_after = [] 
xgb_tuning_train_score_diff_with_tuning = []
xgb_tuning_test_score_diff_before_after_tuning = []

# Loop through each resampling method and each model
for name, resample in best_resamples:
    # Model pipeline without hyperparameter tuning
    model_pipeline_before_tuning = ImbPipeline([
        ('transformer', transformer),
        ('resampling', resample),
        ('model', xgb)
    ])
    
    # Fit the model pipeline without tuning
    model_pipeline_before_tuning.fit(X_train, y_train)
    # Predict on the test set before tuning
    y_pred_test_before = model_pipeline_before_tuning.predict(X_test)
    # Calculate F2 score before tuning
    f2_score_test_before = fbeta_score(y_test, y_pred_test_before, beta=2)
    
    
    # Initialize RandomizedSearchCV for hyperparameter tuning
    random_search = RandomizedSearchCV(
        estimator=model_pipeline_before_tuning, 
        param_distributions=hyperparam_space, 
        n_jobs=-1,
        n_iter=50, 
        scoring=make_scorer(fbeta_score, beta=2), 
        cv=skfold, 
        random_state=RANDOM_STATE
    )
    
    # Fit RandomizedSearchCV for hyperparameter tuning
    random_search.fit(X_train, y_train)
    
    # Get the best estimator from RandomizedSearchCV
    best_model = random_search.best_estimator_
    
    # Predicting on the training set
    y_pred_train = best_model.predict(X_train)
    
    # Predicting on the test set after tuning
    y_pred_test_after = best_model.predict(X_test)
    
    # Evaluate the scores
    f2_score_train = fbeta_score(y_train, y_pred_train, beta=2)
    f2_score_test_after = fbeta_score(y_test, y_pred_test_after, beta=2)
    
    # Calculate the score differences
    train_score_diff_with_tuning = f2_score_train - f2_score_test_after
    
    # Append results to lists
    resampling_names.append(name)
    xgb_tuning_train_score.append(f2_score_train)
    xgb_tuning_test_score_after.append(f2_score_test_after)
    xgb_tuning_test_score_before.append(f2_score_test_before)
    xgb_tuning_train_score_diff_with_tuning.append(train_score_diff_with_tuning)
```

```python

# Predict probabilities
y_prob = best_model.predict_proba(X_test)[:, 1]

th_values = [i/100 for i in range(20, 80)]
score_list = []
best_score = 0
best_th = 0

for th in th_values:
    # Predict with current threshold
    y_pred = np.where(y_prob > th, 1, 0)
    
    # Calculate F2 score
    score_temp = fbeta_score(y_test, y_pred, beta=2)
    score_list.append(score_temp)
    
    # Update best score and threshold if current score is higher
    if score_temp > best_score:
        best_score = score_temp
        best_th = th
print('Best F2 Score:', best_score)
print('Best Threshold:', best_th)
```
